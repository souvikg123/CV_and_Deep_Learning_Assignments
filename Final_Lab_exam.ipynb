```
import numpy as np
import matplotlib.pyplot as plt
import tqdm
import tensorflow as tf
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import ListedColormap
```

```

X = np.array([[2,3,1],
              [1,2,2],
              [2,2,2],
              [3,2,2],
              [5,3,1],
              [6,2,2],
              [7,3,1],
              [8,2,2],
              ])
y = np.array([1,
              1,
              1,
              1,
              0,
              0,
              0,
              0
              ])

Y = np.array([np.eye(2)[i] for i in y])
```

```
X.shape, Y.shape
```

```
from tensorflow.keras import Sequential, Input
from tensorflow.keras.layers import Dense
```

```
model = Sequential([
                    Input(X.shape[1:]),
                    Dense(8, activation='relu'),
                    Dense(2, activation='sigmoid')

])
```

### Before Training

```
x1 = np.linspace(0, 8, 100)
x2 = np.linspace(1, 4, 100)
xx1, xx2 = np.meshgrid(x1,x2)
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.scatter(X[0:4][0], X[0:4][1], X[0:4][2], 'r')
ax.scatter(X[4:][0], X[4:][1], X[4:][2], 'b')
```

```
model.predict(X).argmax(axis=1)
```

```
x1 = np.linspace(-3.0, 3.0, 100)
x2 = np.linspace(-3.0, 3.0, 100)
```

```
model.compile(optimizer='adam', loss='mse', metrics=['accuracy'])
```

```
model.fit(X,Y, epochs=200, verbose=True)
```

```
model.predict(X).argmax(axis=1)
```

```
from sklearn.metrics import classification_report
```

```
print(classification_report(Y.argmax(axis=1), model.predict(X).argmax(axis=1)))
```
